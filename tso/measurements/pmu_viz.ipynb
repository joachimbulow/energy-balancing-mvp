```python
import json
import matplotlib.pyplot as plt
```

```python
with open('pmu_new.json', 'r') as file:
    data = json.load(file)
```

```python
with open('pmu_measurements.json', 'r') as file:
    data = json.load(file)
```

```python
locations = set([d['location'] for d in data])

entries = 1000

for location in locations:
    timestamps = [d['timestamp'] for d in data if d['location'] == location]
    frequencies = [d['frequency'] for d in data if d['location'] == location]
    #if (location == 'Odense'):
    plt.plot(timestamps[:entries], frequencies[:entries], label=location)

plt.xlabel('Timestamp')
plt.ylabel('Frequency')
plt.title('Frequency by Location')
plt.legend()
plt.show()
```
