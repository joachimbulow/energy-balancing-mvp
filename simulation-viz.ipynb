```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import csv


def getSimulationData(file_path):    
    data = []
    with open(file_path, 'r') as csvfile:
        reader = csv.reader(csvfile)
        
        for row in reader:
            # Convert numeric values to appropriate types if needed
            values = row
            time = int(values[0])
            batteryPods = int(values[1])
            nBatteries = int(values[2])
            averageFrequency = float(values[3])
            medianFrequency = float(values[4])
            frequencyStdDev = float(values[5])
            maxFrequency = float(values[6])
            minFrequency = float(values[7])
            chargeSum = int(values[8])
            dischargeSum = int(values[9])
            minLatency = int(values[10])
            maxLatency = int(values[11])
            meanLatency = float(values[12])
            request_interval = float(values[13])

            
            # Create a tuple with the values and append it to the data list
            data.append((time, batteryPods, nBatteries, averageFrequency, medianFrequency, frequencyStdDev,
                         maxFrequency, minFrequency, chargeSum, dischargeSum, minLatency, maxLatency, meanLatency, request_interval))
    
        return data
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

def plotSimulation(file_path):
    data = getSimulationData(file_path)
    battery_n_batteries = [(d[1] * d[2]) for d in data]
    average_frequency = [d[3] for d in data]
    frequency_std_dev = [d[5] for d in data]
    max_frequency = [d[6] for d in data]
    min_frequency = [round(d[7], 6) for d in data]
    charge_actions = [d[8] for d in data]
    discharge_actions = [d[9] for d in data]
    min_latency = [d[10] for d in data]
    max_latency = [d[11] for d in data]
    mean_latency = [d[12] for d in data]
    request_interval = [d[13] for d in data]
    
    primaryColor = '#1976D2'  # Define the color
    secondaryColor = '#FFC107'  # Define the color
    
    plt.rcParams['figure.facecolor'] = '#F5F5F5'  # Set the facecolor of the figure canvas to light gray
    
    # Plot charge and discharge actions
    plt.figure(figsize=(10, 6))
    plt.plot(battery_n_batteries, charge_actions, label='Charge', color=primaryColor, linestyle='--')
    plt.plot(battery_n_batteries, discharge_actions, label='Discharge', color=secondaryColor, linestyle='--')
    plt.xlabel('Total number of batteries')
    plt.ylabel('Number of actions')
    plt.title('Charge and Discharge actions')
    ax = plt.gca()
    ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.0f'))
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.legend()
    plt.show()
    
    # Plot Average Frequency
    plt.figure(figsize=(10, 6))
    plt.plot(battery_n_batteries, average_frequency, color=primaryColor)
    plt.xlabel('Total number of batteries')
    plt.ylabel('Average Frequency (Hz)')
    plt.title('Average Frequency')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
    
    # Plot Frequency Std Dev
    plt.figure(figsize=(10, 6))
    plt.plot(battery_n_batteries, frequency_std_dev, color=primaryColor)
    plt.xlabel('Total number of batteries')
    plt.ylabel('Frequency Std Dev')
    plt.title('Frequency Std Dev')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
    
    # Plot Max Frequency
    plt.figure(figsize=(10, 6))
    plt.plot(battery_n_batteries, max_frequency, color=primaryColor)
    plt.xlabel('Total number of batteries')
    plt.ylabel('Max Frequency (Hz)')
    plt.title('Max Frequency')
    ax = plt.gca()
    ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.6f'))
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

    # Plot Min Frequency
    plt.figure(figsize=(10, 6))
    plt.plot(battery_n_batteries, min_frequency, color=primaryColor)
    plt.xlabel('Total number of batteries')
    plt.ylabel('Min Frequency (Hz)')
    plt.title('Min Frequency')
    ax = plt.gca()
    ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.6f'))
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

    # Plot min latency
    plt.figure(figsize=(10, 6))
    plt.plot(battery_n_batteries, min_latency, color=primaryColor)
    plt.xlabel('Total number of batteries')
    plt.ylabel('Min Latency (ms)')
    plt.title('Min Latency')
    ax = plt.gca()
    ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.0f'))
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

    # Plot max latency
    plt.figure(figsize=(10, 6))
    plt.plot(battery_n_batteries, max_latency, color=primaryColor)
    plt.xlabel('Total number of batteries')
    plt.ylabel('Max Latency (ms)')
    plt.title('Max Latency')
    ax = plt.gca()
    ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.0f'))
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

    # Plot mean latency
    plt.figure(figsize=(10, 6))
    plt.plot(battery_n_batteries, mean_latency, color=primaryColor)
    plt.xlabel('Total number of batteries')
    plt.ylabel('Mean Latency (ms)')
    plt.title('Mean Latency')
    ax = plt.gca()
    ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.0f'))
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

     # Plot mean latency
    plt.figure(figsize=(10, 6))
    plt.plot(request_interval, max_frequency , color=primaryColor)
    plt.xlabel('Request interval (ms)')
    plt.ylabel('Max frequency')
    plt.title('Max frequency')
    ax = plt.gca()
    ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.9f'))
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
```

```python
file_path = './simulations6.csv'
plotSimulation(file_path)
```
