```python
import csv
import matplotlib.pyplot as plt
from dateutil import parser
import matplotlib.dates as mdates
```

```python
def plot_frequency(csv_file, batteries):
    # Lists to store timestamps and frequency values
    timesteps = []
    timestamps = []
    frequencies = []
    first_timestamp = None

    # Read the CSV file and extract the data
    with open(csv_file, 'r') as file:
        reader = csv.reader(file, delimiter=' ')
        # skipping first three lines
        next(reader)
        next(reader)
        next(reader)

        for row in reader:
            timestamp_str = row[0]
            frequency_str = row[1]

            # Skip rows without frequency values
            if frequency_str:
                timestamp = parser.isoparse(timestamp_str)
                timestamps.append(timestamp)
                frequencies.append(float(frequency_str))
                
                if first_timestamp is None:
                    first_timestamp = timestamp

                # Calculate the time difference in seconds and append to timesteps
                timestep = (timestamp - first_timestamp).total_seconds()
                timesteps.append(timestep)
        first_timestamp = None

    # Plot the data
    plt.plot(timesteps[150:200], frequencies[150:200], label=f'{batteries} batteries')
```

```python
def plot_over_timestamp(timestamps, frequencies):
    plt.plot(timestamps, frequencies)
    title = 'Frequency over Time'
    ax = plt.gca()
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))

    plt.xlabel('Timestamp')
    plt.ylabel('Frequency')
    plt.title(title)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
```

```python
def show_timestep_plot(title='Frequency over timesteps'):
    plt.xlabel('Timestep')
    plt.ylabel('Frequency')
    plt.legend()
    plt.title(title)
    plt.grid(True)
    plt.show()
```

```python
# Path to the CSV file
simulation_number = 5

pods = 0
routines = 0
batteries = pods * routines
csv_file = f'simulations{simulation_number}_frequency_{pods}_{routines}.csv'
title = f'Frequency over time for {batteries} batteries'
plot_frequency(csv_file, batteries)


pods = 100
routines = 100
batteries = pods * routines
csv_file = f'simulations{simulation_number}_frequency_{pods}_{routines}.csv'
title = f'Frequency over time for {batteries} batteries'
plot_frequency(csv_file, batteries)

pods = 200
routines = 200
batteries = pods * routines
csv_file = f'simulations{simulation_number}_frequency_{pods}_{routines}.csv'
title = f'Frequency over time for {batteries} batteries'
plot_frequency(csv_file, batteries)

pods = 400
routines = 400
batteries = pods * routines
csv_file = f'simulations{simulation_number}_frequency_{pods}_{routines}.csv'
title = f'Frequency over time for {batteries} batteries'
plot_frequency(csv_file, batteries)

pods = 400
routines = 800
batteries = pods * routines
csv_file = f'simulations{simulation_number}_frequency_{pods}_{routines}.csv'
title = f'Frequency over time for {batteries} batteries'
plot_frequency(csv_file, batteries)

pods = 400
routines = 1000
batteries = pods * routines
csv_file = f'simulations{simulation_number}_frequency_{pods}_{routines}.csv'
title = f'Frequency over time for {batteries} batteries'
plot_frequency(csv_file, batteries)

show_timestep_plot()
```
